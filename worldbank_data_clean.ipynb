# Limpeza de Dados

```python
import wbgapi as wb
```

```python
def read_data(code):
    worldbank_data = wb.data.DataFrame(code, ['DEU', 'AUS', 'CAN', 'USA', 'FRA', 'ITA', 'JPN', 'GBR', 'ZAF', 'SAU', 'ARG', 'BRA', 'CHN', 'KOR', 'IND', 'IDN', 'MEX', 'RUS', 'TUR'])
    return worldbank_data
```
